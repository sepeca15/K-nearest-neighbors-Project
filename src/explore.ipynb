```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movies = pd.read_csv('../data/tmdb_5000_movies.csv')
credit_s = pd.read_csv('../data/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credit_s, on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum()
```

```python

movies.dropna(inplace = True)
```

```python
# genres that are included in the first row

movies.iloc[0].genres
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# converting columns to obtain only the genres without a json format

movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#  function for the 'cast' column for converting

def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i['name'])
        count +=1  
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# how the dataset is coming along

movies.head(1)
```

```python
# From the crew we only obtain the the name of the director

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# overview, convert it in a list by using 'split()' method

movies.overview[0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
# For the recommender system to do not get confused we will remove spaces between words

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# apply function to the 'genres', 'cast', 'crew' and 'keywords' columns

movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# Reduction of the dataset by combining all our previous converted columns into only one column named 'tags'

movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
pd.options.mode.chained_assignment = None  # default='warn'

new_df = movies[['movie_id','title','tags']]

new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df.to_csv('../data/processed/df_processed.csv')
```

## KNN algorithm

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 ,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# 5000 most frequently used words

cv.get_feature_names()
```

```python
# find the cosine_similarity among the movies

from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse =True , key = lambda x:x[1])[1:6]
```

```python
# recommendation function based on the cosine_similarity

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] ##fetching the movie index
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate( distances)),reverse =True , key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
